Import the libraries

```python
import pandas as pd
```

```python
import pyarrow as pa
```

```python
import pyarrow.parquet as pq
```

Define the CSV file path and the Parquet output path

```python
csv_file_path = 'input.csv'
parquet_file_path = 'output.parquet'
```

Load CSV data into a Pandas DataFrame

```python
df = pd.read_csv("data/covid_death2020.csv")
```

print out the head(field) of the csv file with the fist 5 rows/records.

```python
df.head()
```

Data validation: Check for required columns and data types

```python
if 'Province/State' not in df.columns or 'Country/Region' not in df.columns:
        raise ValueError("Missing required columns in CSV")
```

Write the DataFrame to a Parquet file

```python
table = pa.Table.from_pandas(df)
pq.write_table(table, parquet_file_path)
```

read the parquet file

```python
parquet_file = "output.parquet"
df = pd.read_parquet(parquet_file, engine='pyarrow')
df
```
